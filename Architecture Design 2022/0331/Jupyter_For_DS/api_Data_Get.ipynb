```python
# 아파트 정보 목록 조회
# 시도별 아파트 정보 조회해서 아파트 코드 목록 저장
# getAptlist

url = 'http://apis.data.go.kr/B552584/RfidFoodWasteServiceNew/getAptlist'
key = '2xIM1IgBqXn+l2lfsQgsg6hNB7bu1CbSvubTbm/9Q42kTa9XZIsQaRPENls43Clf8PVz9/e3RRH4Uho6Zl6ijQ=='

for city in CityCodeList:
    params ={'serviceKey' : key, 'type' : 'json', 'cityCode' : city,  'rowNum' : 180}
    response_Aptlist = requests.get(url, params=params)
    
    if CityCodeList.index(city) == 0:
        df_Aptlist = pd.DataFrame(response_Aptlist.json()['data']['list'])
        print(CityCodeList.index(city))
    else:
        temp_Aptlist = pd.DataFrame(response_Aptlist.json()['data']['list'])
        
        df_Aptlist = pd.concat([df_Aptlist, temp_Aptlist])
        
        print(CityCodeList.index(city))
# print(response.content)
# print(response_Aptlist.json())

print(df_Aptlist)

    
```
