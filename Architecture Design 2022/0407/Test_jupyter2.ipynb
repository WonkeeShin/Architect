```python
aptCodes = np.unique(csv_test.aptCode)
```

```python
for i in aptCodes:
    sevenlist = [1, 2, 3, 4, 5, 6, 7]
    disdaylist = csv_test[csv_test.aptCode==i].disDay.to_list()
    indexlist = csv_test[csv_test.aptCode==i].disDay.index.to_list()
    
    for n in range(7):
        if n+1 in disdaylist:
            indexlist.remove(indexlist[disdaylist.index(n+1)])
            disdaylist.remove(n+1)
            sevenlist.remove(n+1)
            
#             print(n)
#             print('indexlist', indexlist)
#             print('disdaylist', disdaylist)
#             print('sevenlist', sevenlist)
    
    for n in range(len(sevenlist)):
        csv_test.iloc[indexlist[n], 3] = sevenlist[n]
        
            
```
